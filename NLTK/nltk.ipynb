#### Concordance
It helps to provide if the word occurs in any of the document or text in this example. Concordance view shows us every occurence of a given word, together with some context.

```python
text1.concordance("monstrous")
```

#### Similar
It gives the words that are similar in the context of the words. It provides the list of words that best match

```python
text2.similar("monstrous", num = 4)
```

#### common_context
Allows to examine the contexts that are shared by two or more words

```python
text2.common_contexts(["monstrous" , "very"])
```

#### dispersion_plot
We can determine the location of words in the text i.e. how many words from the begining it appears. When the desired word appears in the text a blue vertical line is plotted, otherwise it remains blank (white line). 

```python
fig = plt.figure(figsize=(12,8))

text4.dispersion_plot(["citizens", "democracy" , "freedom" , "duties" , "America"])
```
