```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv' , engine = 'python',error_bad_lines=False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
X = df.drop('label' , axis = 1)
y = df['label']
```

```python
messages = X.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
messages.reset_index(inplace = True)
```

```python
messages.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(len(messages)):
  review = re.sub('[^a-zA-Z]' , ' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
```

```python
voc_size = 5000
```

```python
onehot_repr=[one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
sent_length = 20

padded_sequence= pad_sequences(onehot_repr, padding = 'pre' , maxlen = sent_length )
padded_sequence
```

```python
vector_sequence = 40
model = Sequential()
model.add(Embedding(voc_size, vector_sequence , input_length = sent_length))
model.add(Bidirectional(LSTM(90)))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics= ['accuracy'])
model.summary()
```

```python
import numpy as np
```

```python
X_final = np.array(padded_sequence)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
model.fit(X_train , y_train , validation_data = (X_test , y_test), epochs = 10 , batch_size = 32)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred= np.where(y_pred>0.6, 1,0)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score

confusion_matrix(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```
